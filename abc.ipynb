```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
url = 'https://vnexpress.net'
uri = "/giao-duc"
driver = webdriver.Chrome(ChromeDriverManager().install())
all_link = []
while True:
     #driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
     time.sleep(2)
     driver.get(url+uri)
     soup = BeautifulSoup(driver.page_source, 'html.parser')
     [all_link.append(y['href']) for y in [x.find("a") for x in soup.find_all('h3', class_='title-news')]]
     next_button = soup.find("a", class_="btn-page next-page")["href"]
     
     uri = next_button
     if uri == "/giao-duc-p21": break
driver.quit()
```

```python
len(all_link)
```

```python
contents_giaoduc = []
count = 0
driver = webdriver.Chrome(ChromeDriverManager().install())
for link in all_link:
  count += 1
  time.sleep(1)
  try:
    driver.get(link)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    time.sleep(1)
    tags_content = soup.find("article", class_= "fck_detail")
    content = "".join([block.text for block in tags_content.find_all("p")])
    if len(content.split()) > 30:
          contents_giaoduc.append(content)
    print(count, len(contents_giaoduc), link)
  except:
    print("failed giao duc", link)
    pass
  if len(contents_giaoduc) > 500: break
```

```python
driver.quit()
```

```python
giaoduc_label = ['giaoduc' for _ in range(len(contents_giaoduc))]
giaoduc_df = pd.DataFrame({"Content":contents_giaoduc, "label": giaoduc_label})
giaoduc_df.to_csv("./data/giaoduc.csv", encoding="utf-8-sig")
giaoduc_df.head()
```
