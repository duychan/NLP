```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
from ipywidgets import IntProgress
from IPython.display import display
import time
```

```python
import json
```

```python
url = 'https://propzy.vn/mua/bat-dong-san/hcm?page='
```

```python
url1 = url
result = requests.get(url1)
content = result.content
soup = BeautifulSoup(content)
```

```python
print(soup.prettify())
```

```python
r = soup.find('script',{'id':'__NEXT_DATA__'}).text
res = json.loads(r)
```

```python
res
```

```python
title = []
bathroom = []
bedroom = []
ward_name = []
district_name = []
street_name = []
price = []
unit_price = []
size = []
direc_name = []
loc_lat = []
loc_lng = []
```

```python
max_count = 10
f = IntProgress(min=0, max=max_count) # instantiate the bar
display(f)
d = 0
for i in range(10):
    url1 = url + str(i)
    result = requests.get(url1)
    content = result.content
    soup = BeautifulSoup(content)
    r = soup.find('script',{'id':'__NEXT_DATA__'}).text
    res = json.loads(r)
    lst = res['props']['initialState']['common']['GET_PROPERTIES_BY_SEARCH']['data']['content']
    # print(len(lst))
    for j in lst:
        title.append(j['title'])
        bathroom.append(j['bathrooms'])
        bedroom.append(j['bedrooms'])
        ward_name.append(j['wardName'])
        district_name.append(j['districtName'])
        street_name.append(j['streetName'])
        price.append(j['priceVND'])
        unit_price.append(j['unitPrice'])
        size.append(j['size'])
        direc_name.append(j['directionName'])
        loc_lat.append(j['location']['lat'])
        loc_lng.append(j['location']['lng'])
    d += len(lst)
    f.value += 1
print(d)
```

```python
df = pd.DataFrame({'Title':title,'Bathrooms':bathroom,'Bedroom':bedroom,'WardName':ward_name,'DistrictName':district_name,
                    'StreetName':street_name,'Price':price,'UnitPrice':unit_price,"Size":size,
                    "Direct":direc_name,'LocLat':loc_lat,'LocLng':loc_lng})
```
